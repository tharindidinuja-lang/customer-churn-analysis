```python
#step:1-setup and data loading
# import libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

#set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
from google.colab import files
uploaded=files.upload()


#mount google drive to access your dataset
#from google.colab import drive
#drive.mount('/content/drive')

#load the data
df=pd.read_csv('/content/Telco_Customer_Churn_Dataset .csv')

#display basic info
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print ("\nMissing values:")
print(df.isnull().sum())

#step2:data cleaning preprocessing
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

#check rows with missing totalcharges
missing_charges=df[df['TotalCharges'].isnull()]
print(f"Customer with missing TotalCharges:{len(missing_charges)}")
print(missing_charges[['customerID','tenure','MonthlyCharges','Churn']])

#fill missing totalcharges with monthlycharges* tenure(for new customers)
df['TotalCharges'].fillna(df['MonthlyCharges'] * df['tenure'], inplace=True);
#convert churn to binary
df['Churn_binary']=df['Churn'].map({'Yes':1,'No':0}) # Note: Changed 'yes'/'no' to 'Yes'/'No' based on typical dataset values. Please verify if 'yes'/'no' are actually lowercase.

#create tenure group
df['tenure_group']=pd.cut(df['tenure'], bins=[0,6,12,24,36,48,72,100], labels=['0-6 months','7-12 months','13-24 months','25-36 months','37-48 months','49-72 months','73-100 months'])
print("\nData cleaning completed")

#create a function to analyze churn by different factors
def analyze_churn_factor(df,factor,figsize=(12,15)):
   fig,axes=plt.subplots(1,2,figsize=figsize)

   #calculate churn rates
   churn_by_factor = df.groupby(factor)['Churn_binary'].agg(['count','mean'])
   churn_by_factor.columns=['count','churn_rate']
   churn_by_factor['churn_rate'] = churn_by_factor['churn_rate'] *100
   churn_by_factor=churn_by_factor.sort_values('churn_rate',ascending=False)

   #bar plot for churn rate
   axes[0].barh(churn_by_factor.index,churn_by_factor['churn_rate'], color='skyblue', edgecolor='black') # corrected 'huurn_by_factor', 'chura_rate', and 'color' to 'skyblue'
   axes[0].set_xlabel('Churn rate(%)')
   axes[0].set_title(f"Churn rate by {factor}")
   axes[0].axvline(x=df['Churn_binary'].mean()*100,color='blue', linestyle='--',label='Overall avg')
   axes[0].legend()

   #count plot
   axes[1].bar(churn_by_factor.index, churn_by_factor['count'], color='skyblue', edgecolor='black') # corrected 'edgcolor'
   axes[1].set_xlabel(factor)
   axes[1].set_ylabel('number of customers')
   axes[1].set_title(f'customer count by {factor}')
   axes[1].tick_params(axis='x', rotation=45)

   plt.tight_layout() # corrected typo 'tight_loayout'
   plt.show()
   return churn_by_factor
#analyse churn by key factors
print("="*50)
print("Key factors influencing churn") # Corrected typo 'kry'
print("="*50)

#1.contract type
print("\n1.churn by contract type:")
contract_churn=analyze_churn_factor(df,'Contract')
print(contract_churn)

#2.tenure group
print("\n2. churn by tenure:")
tenure_churn=analyze_churn_factor(df,'tenure_group')
print(tenure_churn)

#3.internet service
print("\n3. churn by internet service:") # Corrected typo 'churen'
internet_churn=analyze_churn_factor(df,'InternetService')
print(internet_churn)

#4. payment method
print("\n4. churn by payment method:") # Corrected typo 'churen'
payment_churn= analyze_churn_factor(df,'PaymentMethod')
print(payment_churn)

#5.senior citizen
print("\n5.churn by senior citizen:")
senior_churn=analyze_churn_factor(df,'SeniorCitizen')
print(senior_churn)

#6. service add-ons impact
service_features=['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'] # Corrected typos 'OnlineScurity', 'OnlineBachup'
print("\n6. churn by service add-ons:") # Corrected typo 'serivec'
service_impact=pd.DataFrame()
for feature in service_features:
   churn_rate=df.groupby(feature)['Churn_binary'].mean()*100
   service_impact[feature]=churn_rate
print(service_impact)

#create correlation heatmap for numerical features
print("\n7. correlation with churn:") # Corrected typo 'corretation'
numerical_cols=['tenure','MonthlyCharges','TotalCharges','SeniorCitizen','Churn_binary'] # Corrected typo 'Chure_binary'
correlation_matrix=df[numerical_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',center=0,square=True, linewidth=1)
plt.title('Correlation Matrix with Churn')
plt.tight_layout() # corrected typo 'tight_loayout'
plt.show()

print("\nCorrelation with Churn:")
print(correlation_matrix['Churn_binary'].sort_values(ascending=False))

#customer lifetime(LTV) values calculation
print("\n"+"="*50) # Added missing equal signs to match header length
print("Customer lifetime value calculation")
print("="*50)

#assume gross margin
gross_margin=0.70
#calculate basic LTV metrics
avg_monthly_charge=df['MonthlyCharges'].mean() # Corrected typo 'MonthlyChurges'
avg_tenure=df['tenure'].mean()
total_customers=len(df)
total_revenue=df['TotalCharges'].sum()

print(f"\nBasic Metrics:") # Corrected typo 'Matrics'
print(f"Avg monthly charge:${avg_monthly_charge:.2f}")
print(f"Avg Tenure:{avg_tenure:.1f} months") # Removed extra colon
print(f"Avg customer:{total_customers:,}")
print(f"Total Revenue:${total_revenue:,.2f}")

#calculate LTV using different methods
print("\n"+"="*30)
print("LTV calculation methods")
print("="*30)

#method1:simple LTV
simple_ltv=(avg_monthly_charge*avg_tenure)*gross_margin
print(f"\nMethod1 - Simple LTV (Avg monthly * Avg Tenure * Margin):") # Corrected string formatting
print(f"LTV=(${avg_monthly_charge:.2f} * {avg_tenure:.1f} * {gross_margin:.0%}) = ${simple_ltv:,.2f}")

#method2:cohort -based LTV
print(f"\nMethod 2 - Cohort-based LTV:") # Corrected string formatting
cohort_ltv=[]
for group in df['tenure_group'].unique():
   cohort_data=df[df['tenure_group']==group]
   cohort_avg_monthly=cohort_data['MonthlyCharges'].mean()
   cohort_avg_tenure=cohort_data['tenure'].mean() # Corrected 'cohore_data' and 'cohort_avg_value' to 'cohort_avg_tenure'
   cohort_ltv_value=(cohort_avg_monthly*cohort_avg_tenure)*gross_margin # Corrected 'cohore_ltv_value'
   cohort_ltv.append({
   'Tenure group':group,
   'Avg monthly':f"${cohort_avg_monthly:.2f}",
   'Avg tenure':f"{cohort_avg_tenure:.1f} months", # Corrected string formatting
   'LTV ':f"${cohort_ltv_value:,.2f}"
   }) # Corrected dictionary and list append syntax

cohort_ltv_df=pd.DataFrame(cohort_ltv) # Corrected typo 'cohore_ltv_df'
print(cohort_ltv_df.to_string(index=False))

#method3: segment based LTV
print(f"\nMethod 3 - Segment-based LTV:") # Corrected string formatting
segment={
   'Fiber optic':df[df['InternetService']=='Fiber optic'], # Corrected typo 'Internetservice'
   'DSL':df[df['InternetService']=='DSL'], # Corrected typo 'InternerServices' and added quotes around 'DSL'
   'Month-to-month':df[df['Contract']=='Month-to-month'], # Corrected typo 'Contrac' and 'Month to month' to 'Month-to-month'
   'Two year contract':df[df['Contract']=='Two year'], # Corrected label
   'High value(month>$80)':df[df['MonthlyCharges']>80],
   'Low value(monthly<40)':df[df['MonthlyCharges']<40]
}

segment_ltv=[]
for segment_name, segment_data in segment.items(): # Renamed 'segment_date' to 'segment_data' for clarity
   if len(segment_data)>0:
      seg_avg_monthly=segment_data['MonthlyCharges'].mean()
      seg_avg_tenure=segment_data['tenure'].mean()
      seg_ltv_value=(seg_avg_monthly*seg_avg_tenure)*gross_margin

      segment_ltv.append({
         'Segment':segment_name,
         'customers':len(segment_data),
         'Avg_monthly':f"${seg_avg_monthly:.2f}",
         'LTV':f"${seg_ltv_value:,.2f}"
      }) # Added missing comma

segment_ltv_df=pd.DataFrame(segment_ltv)
print(segment_ltv_df.to_string(index=False)) # Corrected typo 'Fales'

#method 4:predictive LTV
print(f"\nMethod 4 - Churn-adjusted LTV:") # Corrected string formatting
churn_rate=df['Churn_binary'].mean() # Corrected typo 'Chuirn_binary'
retention_rate=1-churn_rate
customer_lifetime_months=1/churn_rate
predictive_ltv=(avg_monthly_charge*customer_lifetime_months)*gross_margin # Corrected typo 'customer_lifetime_months_months'

print(f"Overall churn rate:{churn_rate:.1%}")
print(f"Retention rate:{retention_rate:.1%}") # Corrected typo 'Rerention'
print(f"Average customer lifespan:{customer_lifetime_months:.1f} months") # Corrected variable name 'customer_lifetime'
print(f"Predictive LTV:${predictive_ltv:,.2f}")

#visualize LTV key segments
fig,axes=plt.subplots(2,2,figsize=(15,10))

#LTV by contract type
contract_ltv=df.groupby('Contract').apply(
   lambda x:(x['MonthlyCharges'].mean()*x['tenure'].mean())*gross_margin # Corrected typo 'lamdba'
)
axes[0,0].bar(contract_ltv.index, contract_ltv.values, color='lightcoral')
axes[0,0].set_title('LTV by contract type') # Corrected 'Ltv'
axes[0,0].set_ylabel('LTV ($)')
for i, v in enumerate(contract_ltv.values):
   axes[0,0].text(i,v,f'${v:,.0f}',ha='center',va='bottom') # Corrected 'Center'


#LTV by internet service
internet_ltv=df.groupby('InternetService').apply(
   lambda x:(x['MonthlyCharges'].mean()*x['tenure'].mean())* gross_margin
)

axes[0,1].bar(internet_ltv.index, internet_ltv.values, color='lightcoral')
axes[0,1].set_title('Ltv by internet service')
axes[0,1].set_ylabel('LTV ($)')
for i, v in enumerate(internet_ltv.values):
   axes[0,1].text(i,v,f'${v:,.0f}',ha='center',va='bottom')

#LTV by payment method
payment_ltv=df.groupby('PaymentMethod').apply(
   lambda x:(x['MonthlyCharges'].mean()*x['tenure'].mean())*gross_margin
).sort_values(ascending=False)
axes[1,0].bar(payment_ltv.index, payment_ltv.values, color='lightcoral')
axes[1,0].set_title('LTV by payment method')
axes[1,0].set_ylabel('LTV ($)')
for i, v in enumerate(payment_ltv.values):
   axes[1,0].text(i,v,f'${v:,.0f}',ha='center',va='bottom')

#LTV by senior citizen status
senior_ltv=df.groupby('SeniorCitizen').apply(
   lambda x:(x['MonthlyCharges'].mean()*x['tenure'].mean())*gross_margin
)
axes[1,1].bar(senior_ltv.index.astype(str), senior_ltv.values, color=['gold','orange'])
axes[1,1].set_title('LTV by senior citizen status')
axes[1,1].set_ylabel('LTV ($)')
for i, v in enumerate(senior_ltv.values):
   axes[1,1].text(i,v,f'${v:,.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

#5. itentify  high value customer at risk of Churning

print("\n"+"="*50)
print("High value customers at risk of Churning")
print("="*50)

#create a risk scoring function
def calculate_churn_risk(row):
   # Calculate churn risk score based on multiple factors
   # High score = higher risk
   risk_score=0
    #contract type(Month to month= high risk)
   if row['Contract']=='Month-to-month':
      risk_score += 30
   elif row['Contract']=='One year':
      risk_score += 10

   #Tenure(new customer= high risk)
   if row['tenure'] < 12:
      risk_score += 25
   elif row['tenure'] < 24:
      risk_score += 15
   elif row['tenure'] <36:
      risk_score += 5

   # payment method(electronic check=high risk)
   if 'Electronic check' in str(row['PaymentMethod']):
      risk_score += 25
   elif 'Mailed check' in str(row['PaymentMethod']):
      risk_score += 10

   #Internet services(fiber optic=high risk)
   if row['InternetService']== 'Fiber optic':
      risk_score += 15

   # Missing services(no security/backup=high risk)
   if row['OnlineSecurity'] in ['No','No internet service']:
      risk_score += 10
   if row['TechSupport'] in ['No','No internet service']:
      risk_score += 10

   #senior citizen
   if row['SeniorCitizen']==1:
      risk_score += 10

   # Monthly charges(high value customer=high priority)
   if row[ 'MonthlyCharges'] >60:
      risk_score += 10
   return risk_score

# calculate risk score for all customers
df['churn_risk_score']=df.apply(calculate_churn_risk, axis=1)

#calculate LTV each customer(simple version)
df['customer_ltv'] =(df['MonthlyCharges']*df['tenure']*gross_margin)

#identify high value customers(top 25% by LTV)
ltv_threshold=df['customer_ltv'].quantile(0.75)
high_value_custometrs= df[df['customer_ltv']>=ltv_threshold]

print(f"\nLTV threshold for high values customer(75%): ${ltv_threshold:,.2f}")
print(f"Number of high value customer:{len(high_value_custometrs):,}")

#find high value customers atrisk(high LTV +high risk score)
risk_threshold= high_value_custometrs['churn_risk_score'].quantile(0.75)
at_risk_high_value=high_value_custometrs[
   (high_value_custometrs['churn_risk_score'] >=risk_threshold) &
   (high_value_custometrs['Churn']=='No')
].copy()

print(f"\nHigh value custoners at risk churning:{len(at_risk_high_value):,}")
print(f"\n Risk score Threshold:{risk_threshold:.0f}")

# sort by risk score and LTV
at_risk_high_value= at_risk_high_value.sort_values(
   ['churn_risk_score','customer_ltv'],
   ascending=[False,False])

#display top 20 at risk high value customers
print("\n"+"-"*80)
print("Top 20 high value at risk customers")
print("="*80)

display_cols=['customerID','MonthlyCharges','tenure','Contract','InternetService','PaymentMethod','customer_ltv',
'churn_risk_score','OnlineSecurity','TechSupport']

top_at_risk=at_risk_high_value[display_cols].head(20)
top_at_risk['customer_ltv']=top_at_risk['customer_ltv'].apply(lambda x:f"${x:,.2f}")
print(top_at_risk.to_string(index=False))

#create a risk matrix visualization
plt.figure(figsize=(10,6))

#plot all high value customers
scatter=plt.scatter(
   high_value_custometrs['churn_risk_score'],
   high_value_custometrs['customer_ltv'],
   c=high_value_custometrs['Churn_binary'],
   cmap='RdYlGn',
   alpha=0.6,
   s=50,
   linewidth=0.5,
   label='High Value Customers',
   edgecolor='black'
)

#highlight at-risk value customers
plt.scatter(
   at_risk_high_value['churn_risk_score'],
   at_risk_high_value['customer_ltv'],
   c='red',
   alpha=0.8,
   s=100,
   linewidth=1,
   edgecolor='black',
   label='At Risk High Value Customers'
)

plt.xlabel('Churn Risk Score')
plt.ylabel('Customer LTV ($)')
plt.colorbar(scatter, label='Churn Status (0=No, 1=Yes)')
plt.title('Risk Matrix for High Value Customers')
plt.axhline(y=risk_threshold, color='red', linestyle='--', label=f'Risk Threshold{risk_threshold:.0f}')
plt.axvline(x=ltv_threshold, color='blue', linestyle='--', label=f'LTV Threshold${ltv_threshold:,.0f}')

plt.legend()
plt.tight_layout()
plt.show()

#segment at riask customer by contract type
print("\n"+"="*50)
print("At risk high value customers by segment")
print("-"*50)

segment_breakdown=at_risk_high_value.groupby('Contract').agg(
   {'customerID':'count',
   'MonthlyCharges':'mean',
   'customer_ltv':'mean',
   'churn_risk_score':'mean'
   }).round(2)

segment_breakdown.columns=['count','Avg Monthly','Avg LTV', 'Avg Risk Score']
segment_breakdown['Avg Monthly']=segment_breakdown['Avg Monthly'].apply(lambda x: f"${x:,.2f}")
segment_breakdown['Avg LTV']=segment_breakdown['Avg LTV'].apply(lambda x:
   f"${x:,.2f}")
print(segment_breakdown)

#create intervention priority list
print("\n"+"-"*50)
print("Intervention priority top 10 customers to contact")
print("-"*50)

priority_customers=at_risk_high_value.nlargest(10,'churn_risk_score')[[
      'customerID','MonthlyCharges','tenure','Contract','InternetService','PaymentMethod','customer_ltv','churn_risk_score'
   ]].copy()

priority_customers['MonthlyCharges']=priority_customers['MonthlyCharges'].apply(lambda x:f"${x:,.2f}")
priority_customers['customer_ltv']=priority_customers['customer_ltv'].apply(lambda x:f"${x:,.2f}")
print(priority_customers.to_string(index=False))

#6.datadriven retention strategies
print("\n"+"="*50)
print("Data driven retention strategies")
def calculate_retention_value(customer_saved,avg_ltv):
   """Calculate value of retaining customers"""
   return customer_saved*avg_ltv

#strategy 1:contract incentive program
print("\n"+"-"*50)
print("strategy 1: contract incentive program")
print("-"*50)

month_to_month_customers=df[df['Contract']=='Month-to-month']
month_to_month_churn=month_to_month_customers['Churn_binary'].mean()*100
avg_month_to_month_ltv=(month_to_month_customers['MonthlyCharges'].mean()* month_to_month_customers['tenure'].mean())*gross_margin

print(f"Month-to-month customers:{len(month_to_month_customers):,}")
print(f"customer churn rate:{month_to_month_churn:.1f}%")
print(f"Avg LTV:${avg_month_to_month_ltv:,.2f}")

#estimate impact of converting 20% of month_to_month to contracts
convert_target = int(len(month_to_month_customers)*0.2)
retention_value_contract_incentive = calculate_retention_value(convert_target, avg_month_to_month_ltv)

print(f"\nTarget: convert{convert_target:,} month-to-month customer to contracts")
print(f"Potential value saved: ${retention_value_contract_incentive:,.2f}")

#stategy 2:payment method migration
print("\n"+"="*50)
print("Payment method migration")

electronic_check = df[df['PaymentMethod'].str.contains('Electronic check',na=False)]
ec_churn = electronic_check['Churn_binary'].mean()*100
avg_ec_ltv=(electronic_check['MonthlyCharges'].mean()* electronic_check['tenure'].mean())*gross_margin

auto_pay=df[~ df['PaymentMethod'].str.contains('Electronic check', na=False)]
auto_pay_churn = auto_pay['Churn_binary'].mean()*100
print(f"Electronic check users:{len(electronic_check):,}")
print(f"Electronic check churn rate:{ec_churn:.1f}%")
print(f"Auto pay users churn rate: {auto_pay_churn:.1f}%")

#estimate impact of migration 30% of electronic check users
migration_target=int(len(electronic_check)*0.3)
retention_value_payment_migration = calculate_retention_value(migration_target, avg_ec_ltv)

print(f"\nTarget: Migrate{migration_target:,} electronic check users to auto-pay")
print(f"Potential value saved: ${retention_value_payment_migration:,.2f}")

#strategy 3: fiber optic enhancement
print("\n"+"="*50)
print("Strategy: 3 Fiber optic customer experience enhancement")
print("-"*50)

fiber_optic = df[df['InternetService']=='Fiber optic']
fiber_churn = fiber_optic['Churn_binary'].mean()*100
avg_fiber_ltv = (fiber_optic['MonthlyCharges'].mean()* fiber_optic['tenure'].mean())*gross_margin

print(f"Fiber optic customer:{len(fiber_optic):,}")
print(f"Fiber optic churn rate:{fiber_churn:.1f}%")
print(f"Avg LTV:${avg_fiber_ltv:,.2f}")


# identify fiber opticcustomers missing key services
fiber_no_security = fiber_optic[fiber_optic['OnlineSecurity']=='No']
fiber_no_tech = fiber_optic[fiber_optic['TechSupport']=='No'] # Corrected typo 'finer_optic'

# For simplicity, let's assume improving 500 fiber optic customers' experience leads to retention of 500 * (1 - fiber_churn_rate) * avg_fiber_ltv
# Assume a retention uplift of 10% for these 500 customers
retention_value_fiber = avg_fiber_ltv * 500 * 0.1 # Example calculation

#Strategy 4: new customer onbording
print("\n"+"-"*50)
print("Strategy 4: new customer onbording")
print("-"*50)

new_customer = df[df['tenure']<=6]
new_customer_churn = new_customer['Churn_binary'].mean()*100
avg_new_ltv =(new_customer['MonthlyCharges'].mean()* new_customer['tenure'].mean())*gross_margin

print(f"New customer(<= 6 months):{len(new_customer):,}")
print(f"New customer churn rate:{new_customer_churn:.1f}%")
print(f"Avg LTV for new customers:${avg_new_ltv:,.2f}")

# estimate impact of reducing new customer churn by 20%
target_churn_reduction=0.2
customer_saved_new = int(len(new_customer)*(new_customer_churn/100)*target_churn_reduction)
potential_savings_new_customer = calculate_retention_value(customer_saved_new, avg_new_ltv)

print(f"\nTarget: Reduce new customer churn by 20%")
print(f"Potential customers saved: {customer_saved_new:,}")
print(f"Potential value saved: ${potential_savings_new_customer:,.2f}")

#strategy 5: senior citizen program
print("\n"+"-"*50)
print("Strategy 5: senior citizen program")
print("-"*50)

senior_citizen_df = df[df['SeniorCitizen']==1] # Renamed variable to avoid conflict with function
senior_churn = senior_citizen_df['Churn_binary'].mean()*100
avg_senior_ltv=(senior_citizen_df['MonthlyCharges'].mean()* senior_citizen_df['tenure'].mean())*gross_margin

print(f"Senior citizens:{len(senior_citizen_df):,}")
print(f"Senior citizen churn rate:{senior_churn:.1f}%")
print(f"Avg LTV for senior citizens:${avg_senior_ltv:,.2f}")

# Assume saving 300 senior citizens through a program
retention_value_senior = calculate_retention_value(300, avg_senior_ltv)

# identify customer with single services
single_service = df[
   (df['PhoneService']=='Yes')&
   (df['InternetService']=='No')|
   (df['PhoneService']=='No')&
   (df['InternetService']!='No')
]

bundle_customer = df[
   (df['PhoneService']=='Yes') &
   (df['InternetService']!='No')
]

single_churn = single_service['Churn_binary'].mean()*100
bundle_churn = bundle_customer['Churn_binary'].mean()*100

print(f"Single services customers:{len(single_service):,}")
print(f"Single services churn rate:{single_churn:.1f}%")
print(f"Bundle customers:{len(bundle_customer):,}")
print(f"Bundle churn rate:{bundle_churn:.1f}%")

# Assume converting 400 single service customers to bundles with an LTV uplift
retention_value_bundle = avg_monthly_charge * 24 * 0.3 * 400 # example for bundle

# chrate strategy comparison chart
strategies ={
   'Contract Incentive':{
      'target':len(month_to_month_customers),
      'potential_impact':retention_value_contract_incentive,
      'implementation_cost':50000,
      'roi':(retention_value_contract_incentive-50000)/50000*100

   },
   'Payment Migration':{
      'target':len(electronic_check),
      'potential_impact':retention_value_payment_migration,
      'implementation_cost':30000,
      'roi':(retention_value_payment_migration-30000)/30000*100
   },
   'Fiber Optic Enhancement':{
      'target':len(fiber_optic),
      'potential_impact':retention_value_fiber,
      'implementation_cost':75000,
      'roi':(retention_value_fiber-75000)/75000*100
   },
   'New customer onboarding':{
      'target':len(new_customer),
      'potential_impact':potential_savings_new_customer,
      'implementation_cost':40000,
      'roi':(potential_savings_new_customer-40000)/40000*100
   },
   'Senior citizen program':{
      'target':len(senior_citizen_df),
      'potential_impact':retention_value_senior,
      'implementation_cost':25000,
      'roi':(retention_value_senior-25000)/25000*100
   },
   'Bundle discount':{
      'target':len(single_service),
      'potential_impact':retention_value_bundle,
      'implementation_cost':35000,
      'roi':(retention_value_bundle-35000)/35000*100
   }
}

strategy_df=pd.DataFrame(strategies).T
strategy_df= strategy_df.sort_values('roi', ascending=False)

print("\n"+"="*50)
print("Retention strategy comparison")
print("="*50)
print(strategy_df.round(2))

#visualize strategy comparison
fig, axes = plt.subplots(1,2, figsize=(15,6))

#ROI comparison
axes[0].barh(strategy_df.index, strategy_df['roi'], color='lightgreen', edgecolor='black')
axes[0].set_xlabel('ROI (%)')
axes[0].set_title('Retention Strategy ROI Comparison')
for i, v in enumerate(strategy_df['roi']):
   axes[0].text(v, i, f'{v:.1f}%', ha='left', va='center')

# Potential impact vs cost
axes[1].scatter(strategy_df['implementation_cost']/1000, strategy_df['potential_impact']/1000000, color='lightblue', edgecolor='black', s=100, alpha=0.6)
for idx, row in strategy_df.iterrows():
   axes[1].annotate(idx,
   (row['implementation_cost']/1000,
   row['potential_impact']/1000000),
   fontsize=9, ha='center')
axes[1].set_xlabel('Implementation Cost ($k)')
axes[1].set_ylabel('Potential Impact ($M)') # Changed label for better readability
axes[1].set_title('Retention Strategy Cost vs Potential Impact')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


#implementation roadmap and recommendations
print("\n"+"="*50)
print("Implementation roadmap")
#create timeline for implementation
timeline = pd.DataFrame({
   'Phase':['Immediate (0-3 months)','Immediate (0-3 months)', 'Critical (0-3 months)',
   'Short-term (3-6 months)', 'Short-term (3-6 months)','Short-term (3-6 months)',
   'Long-term (6-12 months)','Long-term (6-12 months)','Long-term (6-12 months)'] ,
   'Strategy':['Payment migration campaign','New customer onboarding','Contact top 500 at-risk customer',
   'Contract Incentive offers','Fiber optic Enhancement','Senior citizen program','Bundle discount strategy',
   'Predictive churn model','Automated Retention workflows'],
   'Priority':['High','High','Critical','Medium','Medium','Low','Low','Medium','Medium'], # Added missing priority
   'Estimated cost':['$30,000','$40,000','$15,000','$50,000','$75,000','$25,000','$35,000','$60,000','$45,000'], # Added missing cost
   'Expected impact':['$1.2M','$2.1M','$1.8M','$2.5M','$1.5M','$0.9M','$1.1M','$1.8M','$1.3M']
})

print("\nImplementation timeline:")
print(timeline.to_string(index=False))
```
